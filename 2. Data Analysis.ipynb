### Cleaning the Voting column

```python
def convert_voting(vote):
    try:
        # Attempt to strip and process the value as a string
        vote = vote.strip()  # Remove leading and trailing spaces
        vote = vote.replace('(', '').replace(')', '')  # Remove parentheses if present
        if 'M' in vote:
            return int(float(vote.replace('M', '')) * 1_000_000)  # Convert millions to integer
        elif 'K' in vote:
            return int(float(vote.replace('K', '')) * 1_000)  # Convert thousands to integer
        return int(vote)  # Convert directly to integer
    except AttributeError:
        # If an AttributeError occurs (e.g., vote is an integer), return the value as is
        return int(vote)
    except ValueError:
        # Handle other cases where conversion fails
        return None

# Apply the conversion function to the 'Voting' column
df['Voting'] = df['Voting'].apply(convert_voting)

# Ensure 'Voting' is converted to an integer type while handling missing values
df['Voting'] = df['Voting'].astype('Int64')
```

### Converting the Duration column

```python
def convert_to_minutes(duration):
    try:
        # Remove leading/trailing whitespace
        duration = duration.strip()
        hours, minutes = 0, 0
        if 'h' in duration:
            parts = duration.split('h')
            hours = int(parts[0].strip())  # Extract hours
            if len(parts) > 1 and 'm' in parts[1]:
                minutes = int(parts[1].replace('m', '').strip())  # Extract minutes
        elif 'm' in duration:
            minutes = int(duration.replace('m', '').strip())  # Handle only minutes
        return hours * 60 + minutes
    except (AttributeError, ValueError):  # Handle invalid types or malformed strings
        return None  # Return None for invalid entries


# Applying the function
df['Duration'] = df['Duration'].apply(convert_to_minutes)




```
