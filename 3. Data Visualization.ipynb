### Distribution Of Ratings

High-Rated Movies:

Low-Rated Movies:

KDE Insights:

4. The KDE curve reveals a smooth density estimation, with one clear peak (modal value) around the most common IMDb rating, highlighting a single dominant range for movie ratings.

### Average Ratings By Year

```python
# Grouping data by 'Year' and calculating the average IMDb rating
average_ratings_by_year = df.groupby('Year')['Imdb_Rating'].mean()

# Plotting the line chart
plt.figure(figsize=(12, 6))
plt.plot(average_ratings_by_year.index, average_ratings_by_year.values, marker='o', color='blue', linestyle='-')

plt.title('Average IMDb Ratings by Year', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average IMDb Rating', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

### Insights from the visualization of Average IMDb Ratings by Year:

1. **Trend Analysis:** The line chart reveals fluctuations in average IMDb ratings over the years, indicating variations in audience reception and movie quality across time periods.

2. **Peaks and Troughs:** Identify specific years with notably high or low average ratings, which could correspond to the release of exceptionally popular or poorly received movies.

3. **Consistency Observation:** Certain time frames show consistent average ratings, reflecting stable audience preferences and movie quality during those years.

4. **Recent Trends:** Analyzing the latest years on the chart provides insight into the current state of cinema, including potential shifts in audience expectations or movie-making trends.

```python

# Create a figure and set of subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# 1. Line Plot
sns.lineplot(data=df, x=df.index, y='Imdb_Rating', ax=axs[0, 0], color='green')
axs[0, 0].set_title('Line Plot of IMDb Ratings')
axs[0, 0].set_xlabel('Index')  # Corrected label for X-axis
axs[0, 0].set_ylabel('IMDb Rating')

# 2. Bar Plot
ratings_count = df['Imdb_Rating'].value_counts().sort_index()
sns.barplot(x=ratings_count.index, y=ratings_count.values, ax=axs[0, 1], color='red', edgecolor='black')
axs[0, 1].set_title('Bar Plot of IMDb Ratings')
axs[0, 1].set_xlabel('IMDb Rating')
axs[0, 1].set_ylabel('Number of Movies')

# 3. Box Plot
sns.boxplot(x=df['Imdb_Rating'], ax=axs[1, 0], color='yellow' )
axs[1, 0].set_title('Box Plot of IMDb Ratings')
axs[1, 0].set_xlabel('IMDb Rating')

# 4. Density Plot
sns.kdeplot(df['Imdb_Rating'], ax=axs[1, 1], color='blue', fill=True)
axs[1, 1].set_title('Density Plot of IMDb Ratings')
axs[1, 1].set_xlabel('IMDb Rating')
axs[1, 1].set_ylabel('Density')

# Adjust layout
plt.tight_layout()
plt.show()
```
